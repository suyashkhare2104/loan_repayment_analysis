### Ohkay, Time to plot some graphs and analyse each column!


```python
def univariate(df,col,vartype,hue =None):
    
    '''
    Univariate function will plot the graphs based on the parameters.
    df      : dataframe name
    col     : Column name
    vartype : variable type : continuos or categorical
                Continuos(0)   : Distribution, Violin & Boxplot will be plotted.
                Categorical(1) : Countplot will be plotted.
    hue     : It's only applicable for categorical analysis.
    
    '''
    sns.set(style="darkgrid")
    
    if vartype == 0:
        df2=df.loc[df['loan_repaid']==0,:]
        df= df.loc[df['loan_repaid']==1,:]
        
        fig, ax=plt.subplots(nrows =1,ncols=3,figsize=(20,8))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[col],ax=ax[0],color='sandybrown')
        sns.distplot(df2[col],ax=ax[0])
        ax[1].set_title("Violin Plot")
        sns.violinplot(data =df, x=col,ax=ax[1], inner="quartile",color='sandybrown')
        sns.violinplot(data =df2, x=col,ax=ax[1], inner="quartile")
        ax[2].set_title("Box Plot")
        sns.boxplot(data =df, x=col,ax=ax[2],orient='v',color='sandybrown')
        sns.boxplot(data =df2, x=col,ax=ax[2],orient='v')
        
    if vartype == 1:
        temp = pd.Series(data = hue)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 6 + 4*len(temp.unique())
        fig.set_size_inches(width , 7)
        ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
            for p in ax.patches:
                ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(loan))), (p.get_x()+0.05, p.get_height()+20))  
        else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.32, p.get_height()+20)) 
        del temp
    else:
        exit
        
    plt.show()
```

### Analysis with Probability of defaulting
Categorical Variables vs Probability of defaulting
The main motive of this use case to find what parameters are impacting the most on Loan Status that is if a applicant will successfully complete the loan term or will default.

So we will be using a new term now 'Probability of defaulting' that will be equal to :

Probability of Defaulting = Number of Applicants who defaulted/ Total No. of Applicants
 
We will calculate this probability w.r.t each column in bivariate analysis & will see how the Probability of defaulting changes with these columns.

We will create a user defined function for this.

#### UDF: crosstab

'crosstab' function will be used to get the summarized table for each column(passed as a variable) along with probability of default w.r.t that column

```python
def crosstab(df,y,col):
    '''
    df : Dataframe
    col: Column Name
    '''
    crosstab = pd.crosstab(df[col], y ,margins=True)
    crosstab['Probability_defaulting'] = round((crosstab[0]/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab
```

#### UDF: bivariate_prob

'bivariate_prob' function will be used to plot count of values for each column(passed as a variable) stacked across 'loan_repaid' along with probability of defaulting as a line chart.

```python
# Probability of charge off
def bivariate_prob(df, y, col,stacked= True):
    '''
    df      : Dataframe
    col     : Column Name
    stacked : True(default) for Stacked Bar
    '''
    # get dataframe from crosstab function
    plotCrosstab = crosstab(df, y, col)
    
    linePlot = plotCrosstab[['Probability_defaulting']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True)
    ax.set_title(df[col].name.title()+' vs Probability of defaulting',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Probability of defaulting',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()
```

```python
for col in loan.columns:
    bivariate_prob(df = loan, y=y ,col =col)
```

#### Alright, now its finally time to train some models

```python
X_train, X_test, Y_train, Y_test = train_test_split(loan, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
# Logistic Regression

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = accuracy_score(Y_test, Y_pred)*100
acc_log
```

```python
coeff_df = pd.DataFrame(loan.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# Support Vector Machines

svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = accuracy_score(Y_test, Y_pred)*100
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = accuracy_score(Y_test, Y_pred)*100
acc_knn
```

```python
# Gaussian Naive Bayes

gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = accuracy_score(Y_test, Y_pred)*100
acc_gaussian
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = accuracy_score(Y_test, Y_pred)*100
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = accuracy_score(Y_test, Y_pred)*100
acc_linear_svc
```

```python
# Stochastic Gradient Descent

sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = accuracy_score(Y_test, Y_pred)*100
acc_sgd
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = accuracy_score(Y_test, Y_pred)*100
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
acc_random_forest = accuracy_score(Y_test, Y_pred)*100
acc_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```

#### Now, lets try a voting classifier of top 3 models.

```python
eclf = VotingClassifier(estimators=[('Support Vector Machines', svc), ('Logistic Regression', logreg), ('Random Forest', random_forest)], n_jobs=-1, voting='hard')
eclf.fit(X_train, Y_train)
Y_pred = eclf.predict(X_test)
acc_voting = accuracy_score(Y_test, Y_pred)*100
acc_voting
```
